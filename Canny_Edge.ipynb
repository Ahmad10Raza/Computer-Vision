# Canny Edge : OpenCV

The Canny Edge Detector is a popular edge detection technique used in computer vision and image processing. It was developed by John F. Canny in 1986 and is widely used due to its ability to accurately detect edges in images while minimizing false positives and false negatives. In OpenCV, you can use the `cv2.Canny()` function to apply the Canny edge detection algorithm to an image.

Here are some key features and steps of the Canny Edge Detector:

1. **Gradient Calculation**:
   - The first step in the Canny edge detection algorithm is to calculate the gradient of the image. This is typically done using a convolution operation with a pair of Sobel kernels (one for the x-direction and one for the y-direction).
   - The gradient magnitude and direction at each pixel are computed.

2. **Non-Maximum Suppression**:
   - Non-maximum suppression is performed to thin the edges and keep only the local maxima in the gradient magnitude.
   - At each pixel, the gradient direction is considered, and if the pixel's gradient magnitude is not the maximum among its neighbors in the gradient direction, it is suppressed (set to zero).

3. **Edge Tracking by Hysteresis**:
   - To distinguish between true edges and noise or weak edges, Canny edge detection uses a technique called "edge tracking by hysteresis."
   - The gradient magnitude image is thresholded to identify pixels with strong edges (high gradient magnitude) and pixels with weak edges (gradient magnitude between two threshold values).
   - Strong edges are considered part of the final edge map, while the status of weak edges depends on whether they are connected to strong edges.
   - Weak edges connected to strong edges are also considered part of the final edge map, while isolated weak edges are discarded.

Here's an example of how to apply the Canny edge detection in OpenCV using Python:

```python
import cv2

# Read an image
image = cv2.imread('input.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Canny edge detection
edges = cv2.Canny(image, threshold1=100, threshold2=200)  # Adjust threshold values as needed

# Display the original image and the detected edges
cv2.imshow('Original Image', image)
cv2.imshow('Canny Edges', edges)

# Wait for a key press and close the windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```

````python
#building trackbar with canny edge

#load image into gray scale
img = cv2.imread("Images/unnamed.jpg")
img = cv2.resize(img,(600,700))
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

def nothing(x):
    pass

cv2.namedWindow("Canny")
cv2.createTrackbar("Threshold", "Canny", 0, 255, nothing)

while True:
    a= cv2.getTrackbarPos('Threshold','Canny')
    
    print(a)
    res = cv2.Canny(img_gray,a,255)
    cv2.imshow("Canny",res)
    k=cv2.waitKey(1) & 0xFF
    if k == 27:
        break

cv2.destroyAllWindows()
````
