# Drawing Shapes on Video

````python
import cv2
import datetime
cap = cv2.VideoCapture(0)  #for warn 0 just pass cv2.CAP_DSHOW
print("for width===",cap.get(cv2.CAP_PROP_FRAME_WIDTH))
print("for height==",cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
cap.set(3, 600)  #here 3 for width
cap.set(4, 800)  #here 4 for height 
print("Width====",cap.get(3))
print("Height===",cap.get(4))

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:

       font = cv2.FONT_HERSHEY_COMPLEX_SMALL
       text = ' Height: ' + str(cap.get(4))+' Width: '+ str(cap.get(3))
       date_data = "Date: "+str(datetime.datetime.now())
       frame = cv2.putText(frame, text, (10, 20), font, 1,
                           (0, 155, 255), 1, cv2.LINE_AA)
       frame = cv2.putText(frame, date_data, (20, 50), font, 1,
                           (100, 255, 255), 1, cv2.LINE_AA)
       #Rectangle - accept parameter(img,start_co,end_co,colot ,thickness)
       cv2.rectangle(frame, (384, 10), (510, 128), (128, 0, 255), 8)
       cv2.imshow('frame', frame)

       if cv2.waitKey(1) & 0xFF == ord('q'):
         break
    else:
        break

cap.release()
cv2.destroyAllWindows()
````

To add text to a webcam stream using OpenCV in Python, you can follow these steps:

1. Import the necessary libraries.
2. Initialize the webcam.
3. Read frames from the webcam.
4. Add text to each frame.
5. Display the modified frame with the text.
6. Exit the program when done.

Here's a Python script to accomplish this:

```python
import cv2

# Initialize the webcam
cap = cv2.VideoCapture(0)

while True:
    # Read a frame from the webcam
    ret, frame = cap.read()

    # Check if the frame was read successfully
    if not ret:
        break

    # Add text to the frame
    text = "Hello, OpenCV!"
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 1
    font_color = (255, 255, 255)  # White color in BGR
    thickness = 2
    position = (20, 50)  # Position to place the text

    cv2.putText(frame, text, position, font, font_scale, font_color, thickness)

    # Display the frame with text
    cv2.imshow('Webcam with Text', frame)

    # Exit the loop if 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```

This script opens your webcam, captures frames, adds text to each frame, and displays the frames with text. The loop continues until you press the 'q' key, at which point it releases the webcam and closes the OpenCV windows.

Make sure you have the OpenCV library installed in your Python environment (`pip install opencv-python`) before running this script.

````python
import cv2

# Initialize the webcam
cap = cv2.VideoCapture(0)

while True:
    # Read a frame from the webcam
    ret, frame = cap.read()

    # Check if the frame was read successfully
    if not ret:
        break

    # Add text to the frame
    text = "Hello, OpenCV!"
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 1
    font_color = (255, 255, 255)  # White color in BGR
    thickness = 2
    position = (20, 50)  # Position to place the text

    cv2.putText(frame, text, position, font, font_scale, font_color, thickness)

    # Display the frame with text
    cv2.imshow('Webcam with Text', frame)

    # Exit the loop if 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
````
