```python
import cv2
import pyautogui
import datetime

# Start recording the screen
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('screen_recording.mp4', fourcc, 20.0, (pyautogui.size()[0], pyautogui.size()[1]))

# Start a loop to record each frame of the screen
while True:
    # Capture the current frame of the screen
    img = pyautogui.screenshot()

    # Encode the frame and write it to the video file
    out.write(cv2.imencode('.jpg', img)[1])

    # If the user presses the 'q' key, stop recording
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Stop recording and release the video file
out.release()

# Get the current date and time
now = datetime.datetime.now()
timestamp = now.strftime('%Y-%m-%d_%H-%M-%S')

# Save the video file in the same directory as the Python script
cv2.imwrite(f'screen_recording_{timestamp}.jpg', img)
```

```python
#Screen Recorder ---

import pyautogui as p
import cv2 as c
import numpy as np

#Create resolution
rs = p.size()

#filename in which we store recording
fn = "/home/blackheart/Documents/DATA SCIENCE/DL/OpenCV/Video/"
#Fix the frame rate
fps = 60.0
fourcc = c.VideoWriter_fourcc(*'XVID')
output = c.VideoWriter(fn,fourcc,fps,rs)

#create recording module
c.namedWindow("LIve_Recording",c.WINDOW_NORMAL)

#Resize the window
c.resizeWindow("Live",(600,400))

while True:
    img = p.screenshot() #image
    f = np.array(img) #convert image into array
    f = c.cvtColor(f,c.COLOR_BGR2RGB)
    output.write(f)
    c.imshow("screenshot", f)
    if c.waitKey(1) == ord("q"):
        break

c.destroyAllWindows()
output.release()  
```

```python
import cv2
import pyautogui
import numpy as np

# Screen recording parameters
screen_width, screen_height = pyautogui.size()
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('screen_recording.avi', fourcc, 20.0, (screen_width, screen_height))

while True:
    # Capture the screen
    screenshot = pyautogui.screenshot()

    # Convert the screenshot to an OpenCV image
    frame = np.array(screenshot)
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Write the frame to the video file
    out.write(frame)
    cv2.resize(out, (screen_width, screen_height))
    # Display the recording on the screen (optional)
    cv2.imshow('Screen Recording', frame)
    
    # Check for user input to stop recording (press 'q' key)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video writer and close the OpenCV window
out.release()
cv2.destroyAllWindows()
```

```python
#Capture Multiple Images and Store in a folder

import cv2

vidcap = cv2.VideoCapture(0)
ret,image = vidcap.read()#READ THE VIDEO



count = 0

while True:
  if ret == True:
      cv2.imwrite("frames\\imgN%d.jpg" % count, image)     # save frame as JPEG file
      vidcap.set(cv2.CAP_PROP_POS_MSEC,(count**100)) #used to hold speed of frane generation
      ret,image = vidcap.read()
      cv2.imshow("res",image)
      print ('Read a new frame:',count ,ret)
      count += 1
      if cv2.waitKey(1) & 0xFF == ord("q"):
          break
          cv2.destroyAllWindows()
  else:
      break

vidcap.release() 
cv2.destroyAllWindows()  
```
