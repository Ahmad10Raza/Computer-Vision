OpenCV provides various functions to draw different shapes on images in Python. Here are some common shapes you can draw using OpenCV along with their corresponding functions:

1. **Draw a Line**:
   ```python
   import cv2
   import numpy as np

   # Create a blank image
   img = np.zeros((512, 512, 3), dtype=np.uint8)

   # Draw a blue line from (0,0) to (511,511) with a thickness of 5 pixels
   cv2.line(img, (0, 0), (511, 511), (255, 0, 0), 5)

   # Display the image
   cv2.imshow('Line', img)
   cv2.waitKey(0)
   cv2.destroyAllWindows()
   ```

2. **Draw a Rectangle**:
   ```python
   import cv2
   import numpy as np

   # Create a blank image
   img = np.zeros((512, 512, 3), dtype=np.uint8)

   # Draw a green rectangle from (100,100) to (400,400) with a thickness of 3 pixels
   cv2.rectangle(img, (100, 100), (400, 400), (0, 255, 0), 3)

   # Display the image
   cv2.imshow('Rectangle', img)
   cv2.waitKey(0)
   cv2.destroyAllWindows()
   ```

3. **Draw a Circle**:
   ```python
   import cv2
   import numpy as np

   # Create a blank image
   img = np.zeros((512, 512, 3), dtype=np.uint8)

   # Draw a red circle with center at (250, 250), radius 100, and a thickness of -1 (filled)
   cv2.circle(img, (250, 250), 100, (0, 0, 255), -1)

   # Display the image
   cv2.imshow('Circle', img)
   cv2.waitKey(0)
   cv2.destroyAllWindows()
   ```

4. **Draw an Ellipse**:
   ```python
   import cv2
   import numpy as np

   # Create a blank image
   img = np.zeros((512, 512, 3), dtype=np.uint8)

   # Draw a white ellipse with center at (256, 256), major axis 150, minor axis 100, angle 30 degrees
   cv2.ellipse(img, (256, 256), (150, 100), 30, 0, 360, (255, 255, 255), -1)

   # Display the image
   cv2.imshow('Ellipse', img)
   cv2.waitKey(0)
   cv2.destroyAllWindows()
   ```

These are just a few examples of shapes you can draw using OpenCV in Python. You can customize the color, thickness, and other parameters as needed for your specific application.
