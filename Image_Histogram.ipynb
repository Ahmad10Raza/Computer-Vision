# Image Histogram : OpenCV

In OpenCV, an image histogram is a graphical representation of the distribution of pixel intensity values within an image. It provides valuable information about the image's overall brightness, contrast, and the distribution of colors or grayscale levels. Histograms are commonly used in image processing for tasks like image enhancement, thresholding, and analysis.

Here are the key points to understand about image histograms in OpenCV:

1. **Pixel Intensity Values**: In grayscale images, the pixel intensity values typically range from 0 (black) to 255 (white), with values in between representing various shades of gray. In color images, histograms can be calculated separately for each color channel (e.g., Red, Green, Blue).

2. **Histogram Bins**: A histogram is created by dividing the range of pixel intensity values into a set of "bins" or "buckets." Each bin represents a range of intensity values, and the histogram shows how many pixels in the image fall into each bin.

3. **Histogram Calculation**: OpenCV provides the `cv2.calcHist()` function to compute histograms for images. You can specify the image(s) for which you want to calculate the histogram and the number of bins.

4. **Histogram Visualization**: Histograms are typically visualized as bar graphs, with the x-axis representing the pixel intensity values and the y-axis indicating the number of pixels (or frequency) in each bin. The shape of the histogram reveals information about image characteristics.

5. **Image Enhancement**: Histogram equalization is a technique used to improve the contrast of an image by redistributing the pixel intensity values. It stretches the histogram to occupy the entire intensity range.

6. **Thresholding**: Histogram analysis can help determine an appropriate threshold for image segmentation tasks. For example, you can find a threshold that separates foreground and background in an image.

Here's a simple example of how to calculate and visualize the histogram of a grayscale image using OpenCV:

```python
import cv2
import matplotlib.pyplot as plt

# Read a grayscale image
image = cv2.imread('input.jpg', cv2.IMREAD_GRAYSCALE)

# Calculate the histogram
hist = cv2.calcHist([image], [0], None, [256], [0, 256])

# Plot the histogram
plt.figure(figsize=(8, 4))
plt.hist(image.ravel(), 256, [0, 256])
plt.xlabel('Pixel Intensity Value')
plt.ylabel('Frequency')
plt.title('Image Histogram')
plt.grid(True)
plt.show()
```

In this example, we read a grayscale image, calculate its histogram using `cv2.calcHist()`, and visualize the histogram using Matplotlib. The resulting histogram graphically shows how the pixel intensity values are distributed in the image.

Image histograms are valuable tools for understanding the characteristics of images and making informed decisions in image processing and computer vision applications.

````python
import cv2
import matplotlib.pyplot as plt

# Read a grayscale image
image = cv2.imread('Images/pexels-pixabay-33045.jpg', cv2.IMREAD_GRAYSCALE)

# Calculate the histogram
hist = cv2.calcHist([image], [0], None, [256], [0, 256])

# Plot the histogram
plt.figure(figsize=(8, 4))
plt.hist(image.ravel(), 256, [0, 256])
plt.xlabel('Pixel Intensity Value')
plt.ylabel('Frequency')
plt.title('Image Histogram')
plt.grid(True)
plt.show()
````

````python
#with color image--------------------

img = cv2.imread("Images/color_balls.jpg")
img = cv2.resize(img,(500,650))
b, g, r = cv2.split(img)
cv2.imshow("img", img)
cv2.imshow("b", b)
cv2.imshow("g", g)
cv2.imshow("r", r)
#Plotting different channel with hist
plt.hist(b.ravel(), 256, [0, 256])
plt.hist(g.ravel(), 256, [0, 256])
plt.hist(r.ravel(), 256, [0, 256])
plt.title("ColorFull Image")
plt.show()

#cal
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.title("ColorFull Image")
plt.plot(hist)
plt.show()


#Gray scale
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
hist = cv2.calcHist([img_gray], [0], None, [256], [0, 256])
plt.plot(hist)
plt.title("Gray Image")
plt.show()


#Histogram equalization is good when  of the image is confined to a particular region.
#It accept gray scale image
equ = cv2.equalizeHist(img_gray)
res = np.hstack((img_gray,equ)) #stacking images side-by-side
cv2.imshow("equ",res)
hist1 = cv2.calcHist([equ], [0], None, [256], [0, 256])
plt.plot(hist1)
plt.title("Equalization")
plt.show()


#CLAHE (Contrast Limited Adaptive Histogram Equalization)
# create a CLAHE object (Arguments are optional).
#It is used to enchance image and also handle noise froom image region
#gray scale imge is required
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1 = clahe.apply(img_gray)
cv2.imshow('clahe',cl1)
hist2 = cv2.calcHist([cl1], [0], None, [256], [0, 256])
plt.plot(hist2)
plt.title("CLAHE")
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
````
