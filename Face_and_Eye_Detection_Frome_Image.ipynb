# Face and Eye Detection From Image: OpenCV

To perform face detection from an image using OpenCV in Python, you can use a pre-trained Haar Cascade Classifier for face detection. OpenCV provides pre-trained Haar Cascade models that can detect faces efficiently. Here's how you can do it:

1. **Install OpenCV**:

   If you haven't already installed OpenCV, you can do so using pip:

   ```bash
   pip install opencv-python
   ```

2. **Load the Image**:

   Load the image in which you want to perform face detection. You can use the `cv2.imread()` function to load an image from a file.

   ```python
   import cv2

   # Load the image from a file
   image = cv2.imread('your_image.jpg')
   ```

3. **Load the Pre-trained Haar Cascade Classifier**:

   You can use OpenCV's pre-trained Haar Cascade Classifier for face detection. You can download the XML file containing the classifier from the OpenCV GitHub repository or use the one provided with OpenCV.

   ```python
   # Load the pre-trained Haar Cascade Classifier for face detection
   face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
   ```

4. **Perform Face Detection**:

   Use the `detectMultiScale()` method of the classifier to detect faces in the image.

   ```python
   # Perform face detection
   faces = face_cascade.detectMultiScale(image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
   ```

   - `scaleFactor`: This parameter controls how much the image size is reduced at each image scale. A smaller value will increase detection time but may increase accuracy.
   - `minNeighbors`: This parameter specifies how many neighbors each candidate rectangle should have to retain it. Higher values may result in fewer detections but with higher quality.
   - `minSize`: This parameter sets the minimum object size. Faces smaller than this size will not be detected.

5. **Draw Rectangles Around Detected Faces**:

   Iterate through the detected faces and draw rectangles around them on the original image.

   ```python
   for (x, y, w, h) in faces:
       cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
   ```

6. **Display the Result**:

   Finally, display the image with the detected faces.

   ```python
   # Display the image with detected faces
   cv2.imshow('Face Detection', image)
   cv2.waitKey(0)
   cv2.destroyAllWindows()
   ```

This code will load an image, detect faces, draw rectangles around the detected faces, and display the result. Make sure to replace `'your_image.jpg'` with the path to your image file.

Adjust the parameters of the `detectMultiScale()` method as needed to achieve the desired trade-off between accuracy and speed for your specific image and application.

````python

#Face Detection using haarcascade file 
import cv2
import numpy
face=cv2.CascadeClassifier("cascades/haarcascade_frontalface_default.xml") #for detecting face
eye = cv2.CascadeClassifier('cascades/haarcascade_eye.xml') #for detecting eyes

image=cv2.imread("Images/a.jpg")
gray= cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) #convert into gray 

#parameters(img,scale_factor[reduce image size],min_neighbour)
faces = face.detectMultiScale(gray,4,4)   #for  faces

for(x,y,w,h) in faces:
    
    image=cv2.rectangle(image,(x,y),(x+w,y+h),(127,0,205),3)
    
    #Now detect eyes
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = image[y:y+h, x:x+w]
    eyes = eye.detectMultiScale(roi_gray,1.2,1)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(255,0,0),2)
    
image = cv2.resize(image,(800,700))
#cv2.imshow("Face Detected",image)
import matplotlib.pyplot as plt
plt.imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows() 
````
