# Image Smoothing Using OpenCV

Smoothing, also known as blurring, is a common image processing technique used to reduce noise and detail in an image. OpenCV provides various methods to perform image smoothing. Here, I'll show you how to use the Gaussian Blur and Median Blur methods for image smoothing in OpenCV using Python:

1. **Gaussian Blur**:

   Gaussian blur is a widely used technique for smoothing images. It applies a convolution operation with a Gaussian kernel to the image.

   ```python
   import cv2

   # Read an image
   image = cv2.imread('input.jpg')

   # Apply Gaussian blur
   blurred_image = cv2.GaussianBlur(image, (5, 5), 0)  # You can adjust the kernel size (e.g., (5, 5)) and the standard deviation (0) as needed.

   # Display the original and blurred images
   cv2.imshow('Original Image', image)
   cv2.imshow('Blurred Image', blurred_image)

   # Wait for a key press and close the windows
   cv2.waitKey(0)
   cv2.destroyAllWindows()
   ```

   In this example, we read an image, apply Gaussian blur using `cv2.GaussianBlur()`, and display both the original and blurred images. You can adjust the kernel size and standard deviation based on the level of smoothing you desire.

2. **Median Blur**:

   Median blur is effective for removing salt-and-pepper noise from images. It replaces each pixel's value with the median value in its neighborhood.

   ```python
   import cv2

   # Read an image
   image = cv2.imread('input.jpg')

   # Apply Median blur
   blurred_image = cv2.medianBlur(image, 5)  # You can adjust the kernel size (e.g., 5) as needed.

   # Display the original and blurred images
   cv2.imshow('Original Image', image)
   cv2.imshow('Blurred Image', blurred_image)

   # Wait for a key press and close the windows
   cv2.waitKey(0)
   cv2.destroyAllWindows()
   ```

   In this example, we read an image, apply Median blur using `cv2.medianBlur()`, and display both the original and blurred images. You can adjust the kernel size based on your requirements.

These are just two of the many image smoothing techniques available in OpenCV. Depending on your specific application and the type of noise or details you want to remove, you may choose different methods or combinations of methods for image smoothing.
